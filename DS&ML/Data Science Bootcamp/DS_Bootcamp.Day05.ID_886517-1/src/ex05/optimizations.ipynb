* loop: write a function that iterates through the dataframe using `for i in range(0, len(df))`, `iloc` and `append()` to a list, assign the result of the function to a new column in the dataframe

```python
%%timeit
def use_loop (df):
    res = []
    for i in range(0, len(df)):
        val = df.iloc[i]["Fines"] / df.iloc[i]["Refund"] * df.iloc[i]["Year"]
        res.append(val)
    df["Calculated_data"] = res
use_loop(df)
```

```python
df
```

* do it using `iterrows()`

```python
%%timeit
def use_iterrows(df):
    res = []
    for index, rows in df.iterrows():
        val = rows["Fines"] / rows["Refund"] * rows["Year"]
        res.append(val)
    df["Calculated_data"] = res
use_iterrows(df)
```

```python
df
```

* do it using `apply()` and lambda function

```python
%%timeit
def use_apply(df):
    df["Calculated_data"] = df.apply(lambda x: x["Fines"] / x["Refund"] * x["Year"], axis=1)

use_apply(df)
```

```python
df
```

* do it using `Series` objects from the dataframe

```python
%%timeit
def use_series(df):
    df["Calculated_data"] = df["Fines"] / df["Refund"] * df["Year"]

use_series(df)
```

```python
df
```

* do it as in the previous subtask but with the method `.values`

```python
%%timeit
def use_values(df):
    df["Calculated_data"] = df["Fines"].values / df["Refund"].values * df["Year"].values

use_values(df)
```

* make a `copy()` of your initial dataframe into another dataframe `optimized`

```python
optimized = df.copy()
optimized
```

* downcast from `float64` to `float32` for all the columns
* downcast from `int64` to the smallest numerical dtype possible
* run `info(memory_usage='deep')` for your new dataframe, pay attention to the Dtype and the memory usage

```python
float_cols = optimized.select_dtypes(include="float64").columns
optimized[float_cols] = optimized[float_cols].astype("float32")
int_cols = optimized.select_dtypes(include="int64").columns
optimized[int_cols] = optimized[int_cols].apply(pd.to_numeric, downcast="integer")
optimized.info(memory_usage="deep")
```

* categories:
    * change the `object` type columns to the type `category`
    * This time, check the memory usage, it probably has a decrease of 2–3 times compared to the initial dataframe

```python
object_cols = optimized.select_dtypes("object").columns
optimized[object_cols] = optimized[object_cols].astype("category")
optimized.info(memory_usage="deep")
```
